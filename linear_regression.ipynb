```python
df = pd.read_csv("./data/melb_data.csv")

# Retirar colunas com alta cardinalidade
df = df.drop(["Suburb", "Address", "SellerG", "Date"], axis=1)

# Muitos valores Nulos
df = df.drop(["BuildingArea", "YearBuilt", "CouncilArea"], axis=1)
df.info()
```

```python
df.corr()
```

```python
# Escolhendo colunas com relaÃ§Ã£o mais alta com price
base1 = df[["Price","Rooms","Bedroom2","Bathroom","Car"]]

# Tratando nulos

base1 = base1.dropna(axis=0)

base1
```

###  Criando o Modelo Machine Learning

```python
Y = base1["Price"]
X = base1[["Rooms","Bedroom2","Bathroom", "Car"]]
```

```python
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(X,Y)
```

```python
# Importando o modelo - Linear Regression
from sklearn.linear_model import LinearRegression

# Instanciando o modelo
reg = LinearRegression()

# Realizando o fit do modelo - treino
reg = reg.fit(x_treino, y_treino)

# Usando o modelo para prever os dados em x_teste

y_regressao = reg.predict(x_teste)
```

```python
import plotly.express as px

fig = px.scatter(x=y_teste, y=y_regressao)
fig.show()
```

```python
# Avaliando o erro quadrÃ¡tico mÃ©dio
from sklearn.metrics import mean_squared_error

erro_quadratico_regressao = mean_squared_error(y_teste, y_regressao)
print(erro_quadratico_regressao)
```

```python
# Avaliando o R quadrado

from sklearn.metrics import r2_score

R_quadrado = r2_score(y_teste, y_regressao)
print(R_quadrado)
```

### Arvore de DecisÃ£o

```python
# Importando o modelo
from sklearn.tree import DecisionTreeRegressor


# Instanciando o modelo
reg2 = DecisionTreeRegressor()

# Realizando o fit do modelo - treino
reg2 = reg2.fit(x_treino, y_treino)

# Usando o modelo para prever os dados em x_teste

y_regressao2 = reg2.predict(x_teste)
```

```python
fig = px.scatter(x=y_teste, y=y_regressao2)
fig.show()
```

```python
# Avaliando o erro quadrÃ¡tico mÃ©dio
from sklearn.metrics import mean_squared_error

erro_quadratico_regressao2 = mean_squared_error(y_teste, y_regressao2)
print(erro_quadratico_regressao2)
```

```python
# Avaliando o R quadrado

from sklearn.metrics import r2_score

R_quadrado = r2_score(y_teste, y_regressao2)
print(R_quadrado)
```
